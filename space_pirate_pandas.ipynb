```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```
print(sns.get_dataset_names())
```

```
dataset = sns.load_dataset('diamonds')
dataset.tail(2)
```

```
dataset.describe()
```

```
sns.scatterplot(data=dataset, x='carat', y='price')
```

```
plt.figure(figsize=(8,8))
sns.scatterplot(data=dataset, x='carat', y='price', linewidth=0, alpha=0.3, hue='color')
```

```
dataset.head()
```

```
dataset['depth_x_table'] = dataset['depth'] + dataset['table']
```

```
dataset.head()
```

```
sns.scatterplot(data=dataset, x='carat', y='depth_x_table', linewidth=0, alpha=0.3)
```

```
dataset['volume'] = dataset['x'] + dataset['y']+ dataset['z']
```

```
dataset.head()
```

```
sns.scatterplot(data=dataset, x='carat', y='volume', linewidth=0, alpha=0.3, hue='price')
```

```
dataset.head()
```

```
sample = dataset.sample(frac=0.05, random_state=1)
```

```
sample.head()
```

```
sns.pairplot(data=sample)
```

```
sns.boxplot(data=dataset, x='carat', y='clarity')
```

```
pv = dataset.pivot_table(index='clarity', columns=['cut'], values='price')
```

```
sns.heatmap(pv)
```
